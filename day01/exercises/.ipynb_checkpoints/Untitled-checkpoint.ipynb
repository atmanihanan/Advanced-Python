```python

import matplotlib.pyplot as plt

# Donnes fournies
x = [1, 4, 7, 10, 13, 16, 19, 21]
y = [6869, 2513,3316,4241,4490,3086,8543,2287]

# Tracer la courbe
plt.figure(figsize=(8, 6))
plt.plot(x, y,  label='Data')
plt.xlabel('Block size')
plt.ylabel('Time (m)')
plt.title('MXM  block version \n Change in the calculation time of MXM as the number of blocks changes')
plt.legend()
plt.grid(True)
plt.show()


```
